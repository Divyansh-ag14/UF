```python
# 1. Read the data from the CSV file into a DataFrame and display the first five rows.
df = pd.read_csv("data/ramen_ratings.csv")
df.head()
```

```python
# 2. Create a histogram for the Stars column using 20 bins to display the Star ratings.

plt.hist(df["Stars"], bins=20)
plt.title("Historgram for Stars")
```

```python
# 3.	Create a KDE plot for the Starts column.
# plot.kde(): https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.kde.html

df["Stars"].plot.kde()
plt.title("KDE for Stars")
```

```python
df.head()
```

```python
# 4. Create a bar plot that shows the average stars/scores by ramen style. Hint: You’ll need to group the data before you plot it.
df.groupby("Style")["Stars"].mean().plot(kind="bar")
plt.yticks(ticks=[i * 0.5 for i in range(11)])
plt.title("Average Stars by Ramen Style")
plt.show()
```

```python
# 5.	Create a bar plot that shows the total number of reviews for each ramen Style.

df.groupby("Style").size().plot(kind="bar")
plt.title("total number of reviews for each ramen Style")
```

```python
# 6.	Create a bar plot that shows the highest score/star given to each ramen Style.
# # What is the most common style of ramen?
df.groupby("Style")["Stars"].max().plot(kind="bar")
plt.title("highest score/star given to each ramen Style")
print(f"Most common ramen style: {df.groupby("Style").size().sort_values(ascending=False).index[0]}")
```

```python
# 7.Create a bar plot to show the Stars by Style, with a Seaborn specific method.

fig, ax = plt.subplots(1,2, figsize=(10,5))

mean_s = df.groupby("Style")["Stars"].mean().reset_index()
max_s = df.groupby("Style")["Stars"].max().reset_index()

sns.barplot(data=mean_s, x="Style", y="Stars", ax=ax[0])
ax[0].set_title("Mean Stars")

sns.barplot(data=max_s, x="Style", y="Stars", ax=ax[1])
ax[1].set_title("Max Stars")
```

```python
# 8.Create a bar plot to show the Stars by Style, with a Seaborn generic method.

fig, ax = plt.subplots(1,2, figsize=(10,5))

mean_s = df.groupby("Style")["Stars"].mean().reset_index()
max_s = df.groupby("Style")["Stars"].max().reset_index()

sns.barplot(data=mean_s, x="Style", y="Stars", ax=ax[0])
ax[0].set_title("Mean Stars")

sns.barplot(data=max_s, x="Style", y="Stars", ax=ax[1])
ax[1].set_title("Max Stars")
```

```python
# 9. Modify the plot you just created in Task 8 to use a custom title and x-axis label.
fig, ax = plt.subplots(1,2, figsize=(10,5))

mean_s = df.groupby("Style")["Stars"].mean().reset_index()
max_s = df.groupby("Style")["Stars"].max().reset_index()

sns.barplot(data=mean_s, x="Style", y="Stars", ax=ax[0])
ax[0].set_title("Mean Stars")
ax[0].set_xlabel("Style")

sns.barplot(data=max_s, x="Style", y="Stars", ax=ax[1])
ax[1].set_title("Max Stars")
ax[1].set_xlabel("Style")
```

```python
# 10.	Use the query() method to filter the data to contain only data for Japan, India, Taiwan and the USA.
#     Plot the stars by style for each country on a subplot with two subplots per row, using Seaborn.

df2 = df.query("Country in ['Japan', 'India', 'Taiwan', 'USA']")
countries = ['Japan', 'India', 'Taiwan', 'USA']

plt.figure(figsize=(10,8))
for i, val in enumerate(countries):
    plt.subplot(2,2, i+1)
    sns.barplot(data=df2[df2["Country"]==val], x="Style", y="Stars")
    plt.title(val)
    plt.tight_layout()
```

```python
# 11. Modify the plot you just created to have a custom title for each subplot, using Seaborn. Hint: use the enumerate() method.
for i, val in enumerate(countries):
    plt.subplot(2,2, i+1)
    sns.barplot(data=df2[df2["Country"]==val], x="Style", y="Stars")
    plt.title(f"stars vs style for {val}")
    plt.tight_layout()

```
