```python
import numpy as numpy
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Read the data from the CSV file into a DataFrame and display the first five rows.

```python
df = pd.read_csv("data/avocado.csv")
df.head()
```

### 2. Create a new DataFrame that contains the total of the Small Bags, Large Bags, and XLarge Bags columns grouped by type, and then display the DataFrame.

```python
bags_df = df.groupby("type")[["Small Bags", "Large Bags", "XLarge Bags"]].sum()
bags_df
```

### 3. Use the grouped data to create a bar plot that shows the number of small, large, and extra-large bags for both types of avocado

```python
bags_df.plot(kind="bar")
```

### 4. Use the original data to create a scatter plot for the Total Volume and AveragePrice columns.

```python
sns.scatterplot(data=df, x="Total Volume", y="AveragePrice")
```

### 5. Create the following plot, with a Seaborn specific method.

```python
sns.scatterplot(data=df, y="Total Volume", x="AveragePrice")
```

### 6. Create the same plot, with a Seaborn generic method.

```python
sns.relplot(kind="scatter",data=df, y="Total Volume", x="AveragePrice")
```

### 7. Create the same plot but with the hue parameter set to year and the dots for the total volume ranging in size from 10 to 100.

```python
sns.relplot(data=df, kind= 'scatter', x="AveragePrice", y="Total Volume", hue="year", size="Total Volume", sizes=(10,100))
```

### 8. Create a plot that looks like the following plot, using Seaborn. (The colors can be different.)


```python
region_df = df[df["region"].isin(["TotalUS", "West", "WestTexNewMexico"])]
sns.barplot(data=region_df, x="year", y="Total Volume", hue="region")
```

### 9. In what year were the most bags of avocado sold? You may try a bar plot

```python
print(f"Year with max bags sold: {df.groupby('year')['Total Bags'].sum().sort_values(ascending=False).index[0]}")
```

```python
df.groupby('year')['Total Bags'].sum().plot(kind="bar")
```
