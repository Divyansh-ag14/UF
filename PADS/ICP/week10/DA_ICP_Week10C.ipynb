```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### 1. Read the data from the CSV file into a DataFrame and display the first five rows.

```python
df = pd.read_csv("data/exams.csv")
df.head()
```

### 2. Create a box plot

```python
df.plot(kind="box")
```

### 3. Create the following scatter plot.


```python
sns.scatterplot(data=df, y="reading score", x="math score")
```

### 4. Group the data by gender and calculate the average scores. Then create a bar plot.

```python
df.groupby("gender")[["math score", "reading score", "writing score"]].mean().plot(kind="bar")
```

### 5. Use the Seaborn catplot() method to create a plot like the one below. To make that easy to do, you can use a count plot.


```python
sns.catplot(data=df, kind="count", x="parental level of education")
```

### 6. Rotate the x labels for the above plot to make them readable.

```python
sns.catplot(data=df, kind="count", x="parental level of education")
plt.xticks(rotation=30)
plt.show()
```

### 7. Create the following scatter plot with a Seaborn specific method that compares the writing score with the reading score. Use a different color to show which students took a test prep course.

```python
sns.scatterplot(data=df, x="reading score", y="writing score", hue="test preparation course")
```

### 8. Create the same plot with a Seaborn general method. Optional: adjust the size of the scatter plot to check the differences.

```python
sns.relplot(kind="scatter", data=df, x="reading score", y="writing score", hue="test preparation course")
```

### 9. What is the relationship between the reading score and the writing score? You may try a scatter plot.

```python
sns.scatterplot(data=df, x="reading score", y="writing score")
```

there is a linear relationship between reading and writing scores. the scores tend to increase in the same direction, meaning that if a student got higher marks in writing test, there is a high probablity they will also score good in reading and vica versa.

### 10. Does the type of lunch appear to have an effect on the average test scores? You may group data by lunch and try a bar plot for average scores

```python
df2 = df.groupby("lunch")[["math score", "reading score", "writing score"]].mean().reset_index()
sns.barplot(data=df2)
```

```python
df2
```

```python
plt.figure(figsize=(8,7))
avg_scores = df.groupby("lunch")[["math score", "reading score", "writing score"]].mean().reset_index()
avg_scores = pd.melt(avg_scores, id_vars=["lunch"], var_name="subject", value_name="average_score")
sns.barplot(x="lunch", y="average_score", hue="subject", data=avg_scores)
```

on an average students getting standard lunch are performing better
